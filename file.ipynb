```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("exam_malpractice_dataset.csv")

df['Gender'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("Gender Distribution")
plt.ylabel("Count")
plt.show()
```

```python
df['MalpracticeType'].value_counts().plot(kind='barh', color='orange')
plt.title("Malpractice Types")
plt.xlabel("Count")
plt.show()
```

```python
df['Department'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.ylabel("")
plt.title("Cases by Department")
plt.show()
```

```python
df['Level'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title("Cases by Level")
plt.xlabel("Level")
plt.ylabel("Count")
plt.show()
```

```python
outcome_counts = df['Outcome'].value_counts()
plt.pie(outcome_counts, labels=outcome_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.title("Outcome Distribution")
plt.show()
```

```python
inv_counts = df['Invigilator'].value_counts().head(10)
plt.scatter(inv_counts.index, inv_counts.values, color='red', s=100)
plt.title("Top 10 Invigilators Reporting Cases")
plt.xlabel("Invigilator")
plt.ylabel("Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
df['ExamYear'] = pd.to_datetime(df['ExamDate']).dt.year
df['ExamYear'].value_counts().sort_index().plot(kind='area', alpha=0.5, color='cyan')
plt.title("Cases per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
penalty_counts = df['Penalty'].value_counts()
markerline, stemlines, baseline = plt.stem(penalty_counts.index, penalty_counts.values, basefmt=" ")
plt.setp(markerline, 'color', 'brown')
plt.setp(stemlines, 'color', 'brown')
plt.title("Penalty Distribution")
plt.xlabel("Penalty Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['DurationMin'], df['NumQuestions'], s=df['NumQuestions']*5, alpha=0.5, color='blue')
plt.title("Duration vs Num of Questions")
plt.xlabel("Duration (min)")
plt.ylabel("Num of Questions")
plt.show()
```
